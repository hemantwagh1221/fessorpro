```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import pandas as pd
import time

# Setup Selenium WebDriver (headless mode optional)
options = Options()
options.add_argument("--headless")  # Optional: Run in headless mode
service = Service()  # Assumes chromedriver is in your PATH
driver = webdriver.Chrome(service=service, options=options)

# Open the Wikipedia page
url = "https://en.wikipedia.org/wiki/NIFTY_50"
driver.get(url)

# Wait for the page to fully load
time.sleep(2)  # You could use WebDriverWait instead of sleep

# Get the page source
html = driver.page_source

# Close the driver
driver.quit()

# Use BeautifulSoup to parse the table
soup = BeautifulSoup(html, "html.parser")
tables = soup.find_all("table", {"class": "wikitable"})

# Convert the first table to a DataFrame
if tables:
    df = pd.read_html(str(tables[0]))[0]
    print(df.tail())
else:
    print("No table found.")
```
